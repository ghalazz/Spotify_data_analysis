**1. Load and Inspect the Dataset**

```python
import pandas as pd
df = pd.read_csv('/content/dataset.csv')
df.head()
```

**Checking dimensions**

```python
  print("Shape:",df.shape)
  df.info()
  df.describe()
```

```python
#null values
print("Missing values:\n",df.isnull().sum())

#duplicate rows
print("Duplicate rows:",df.duplicated().sum())
```

**Removing Nulls and unwanted**

```python
df = df.drop('Unnamed: 0',axis=1)
df.head()
```

```python
df = df.dropna(subset=['artists','album_name','track_name'])
print("Missing values:\n",df.isnull().sum())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Exploratory Visualizations**

```python
#Top 10 Most Common Genres
top_genres = df['track_genre'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='crest')
plt.title("Top 10 Track Genres")
plt.xlabel("Number of Tracks")
plt.ylabel("Genre")
plt.show()
```

```python
print(df['track_genre'].value_counts().head(10))
print(f"\nTotal genres: {df['track_genre'].nunique()}")
print(f"Dataset size: {len(df)}")
```

```python
# Top 10 Most Frequent Artists
top_artists = df['artists'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_artists.values, y=top_artists.index, palette='magma')
plt.title("Top 10 Artists by Track Count")
plt.xlabel("Number of Tracks")
plt.ylabel("Artist")
plt.show()
```

```python
# Popularity Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['popularity'], bins=30, color='skyblue')
plt.title("Distribution of Track Popularity")
plt.xlabel("Popularity Score (0–100)")
plt.ylabel("Number of Tracks")
plt.show()
```

```python
# Convert milliseconds to minutes
df['duration_minutes'] = df['duration_ms'] / 60000

# Plot duration
plt.figure(figsize=(8,5))
sns.histplot(df['duration_minutes'], bins=40, color='teal')
plt.title("Distribution of Track Duration (in Minutes)")
plt.xlabel("Duration (minutes)")
plt.ylabel("Count")
plt.show()
```

```python
# Correlation heatmap of numerical audio features
plt.figure(figsize=(12,10))
corr = df[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
           'instrumentalness', 'liveness', 'valence', 'tempo', 'popularity']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Audio Features")
plt.show()
```

```python
# Heatmap of Danceability vs Energy
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='danceability', y='energy', alpha=0.4)
plt.title("Danceability vs Energy")
plt.xlabel("Danceability")
plt.ylabel("Energy")
plt.show()
```

**Final Summary: Exploratory Data Analysis on Spotify Dataset**


Dataset Overview

*   Total tracks: 113,999
*  Features analyzed: genre, artist, popularity, duration, and audio attributes like danceability, energy, loudness, valence, etc.
*   Data source: Kaggle - Spotify Tracks Dataset by Maharshi Pandya









**Key Findings**

---
Top Genres
*  All top 10 genres (e.g., acoustic, alt-rock, ambient, etc.) are evenly distributed with around 1,000 tracks each, indicating a balanced genre representation.

---

Most Frequent Artists

*  Artists like The Beatles, George Jones, and Stevie Wonder appear most frequently, each with over 200 tracks. Showing their broad discography presence in the dataset.

---

Track Popularity

*   Most tracks have a low popularity score, with a sharp peak at 0.
*   Very few tracks reach a popularity score above 80, indicating only a small fraction of songs are considered widely popular.

---

Track Duration

*   Most songs fall between 2–5 minutes, which aligns with common radio play standards.
*   Outliers with very high duration (over 10 mins) exist but are rare.

---

Audio Feature Correlations
*   Energy and loudness show a strong positive correlation (0.76) .Loudertracks are usually more energetic.
*   Danceability and valence also show a moderate positive correlation (0.48), happier songs are often more danceable.
*   Popularity has very weak correlation with any feature, suggesting that external factors (e.g., marketing, artist fame) affect popularity more than audio characteristics.
---

Danceability vs Energy
*   Most songs fall within the 0.4 to 0.8 range for both danceability and energy.
*   Some dense clustering suggests a common “sweet spot” in audio traits for popular music.
